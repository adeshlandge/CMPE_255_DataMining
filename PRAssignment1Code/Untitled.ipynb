```python
dataList = list()
with open('data/format.dat', 'r') as fileData:
    datalines = fileData.readlines();
    totalrows = len(datalines) 
    for i in range(totalrows):
        p = datalines[i].split('\n')
        dataList.append(p)
    
print(dataList)
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score

txt_file = open('data/format.dat', 'r')
format_data = list(txt_file.read())
#print(format_data)

txt_file2 = open('data/outputFinal.dat', 'r')
output_data = list(txt_file2.read())
#print(output_data)

print(normalized_mutual_info_score(format_data,output_data))
```
