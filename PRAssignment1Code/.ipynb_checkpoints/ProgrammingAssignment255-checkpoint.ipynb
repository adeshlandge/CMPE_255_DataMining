```python
import numpy as np
from scipy.sparse import csr_matrix
from sklearn import metrics
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline
import sys
from sklearn.utils import shuffle
from sklearn.metrics import silhouette_score 
```

```python
#Convert sparse data into usable csr format
colList = list()
rowList = list()
dataList = list()
with open('data/train.dat', 'r') as fileData:
    datalines = fileData.readlines();
    totalcolumns = 0;
    shapeCount = 0 # variable to calculate the shape size
    totalrows = len(datalines)  
   
    for i in range(totalrows):
        p = datalines[i].split()
        if len(p) % 2 != 0:
            raise ValueError("Total no.of keys mismatches the total no.of values. It is an invalid compressed sparse row matrix.")
        shapeCount += len(p)//2 #Rounding off to its previous integer value to avoid decimals while calculating shape size.
        for j in range(0, len(p), 2): 
            cid = int(p[j]) - 1
            if cid+1 > totalcolumns:
                totalcolumns = cid+1

val = np.zeros(shapeCount, dtype=np.single) #
ind = np.zeros(shapeCount, dtype=np.intc) #
ptr = np.zeros(totalrows+1, dtype=np.longlong) #
n = 0  #
for i in range(totalrows):
    p = datalines[i].split()
    for j in range(0, len(p), 2): 
        ind[n] = int(p[j]) - 1
        val[n] = float(p[j+1])
        n += 1
    ptr[i+1] = n 
    
assert(n == shapeCount)
    
cm=csr_matrix((val, ind, ptr), shape=(totalrows, totalcolumns), dtype=np.single)
```

```python
#Compute tfidf on the resulting csr_matrix
from sklearn.feature_extraction.text import TfidfTransformer
tdidfTransformer = TfidfTransformer(norm='l2', use_idf=True, smooth_idf=False)
#tfidf = tdidfTransformer.fit_transform(cm)
sparse_matrix = tdidfTransformer.fit_transform(cm).toarray();
```

```python
%%time
#Perform Truncated SVD to reduce the number of columns to 220 ->Dimensionality reduction
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=220,n_iter=7)
X = svd.fit_transform(sparse_matrix)
```

```python

def calculateSimilarity(matrix, centroids):
    #Calculating dot product of the given matrices. A*Transpose(B) to check the similarity between them.
    checkSimilarity = matrix.dot(centroids.T) 
    return checkSimilarity
```

```python
def getCentroidsFromMatrix(matrix):
    #shuffling the matrix to randomly fetch two centroids for the given matrix.
    shuffledMatrix = shuffle(matrix, random_state=0)
    return shuffledMatrix[:2,:]
```

```python
def divideIntoTwoClusters(matrix, centroids):
    matrixSimilarity = calculateSimilarity(matrix, centroids)  # We get the similarity of a matrix by computing a.b = abcos(0)
    firstCluster = list()
    secondCluster = list()
    
    for i in range(matrixSimilarity.shape[0]):
        row = matrixSimilarity[i]
        #sort the matrix in ascending order and get the last value which indicates the value with highest similarity.
        sortedMatrix = np.argsort(row)[-1]        
        if sortedMatrix == 0:
            firstCluster.append(i)
        else:
            secondCluster.append(i)
        
    return firstCluster, secondCluster
```

```python
def recalculateCentroid(matrix, clusters):
    centroidList = list()    
    for i in range(0,2):
        clusterData = matrix[clusters[i],:]
        meanOfCluster = clusterData.mean(0)
        centroidList.append(meanOfCluster)
        
    centroidArray = np.asarray(centroidList)
    
    return centroidArray
```

```python
def kmeans(matrix, numberOfIterations):
    
    centroids = getCentroidsFromMatrix(matrix)
    
    for index in range(numberOfIterations):   #_     
        clustersList = list()        
        firstCluster, secondCluster = divideIntoTwoClusters(matrix, centroids)        
        if len(firstCluster) > 1:
            clustersList.append(firstCluster)
        if len(secondCluster) > 1:
            clustersList.append(secondCluster)            
        centroids = recalculateCentroid(matrix, clusters)
        
    return firstCluster, secondCluster
```

```python
#calculate SSE to measure the amount of variation within a cluster.
def calculateSSE(matrix, clusters):
    sseArray = [];
    sseList = list();    
    
    for clusterVar in clusters:
        members = matrix[clusterVar,:]
        SSE = np.sum(np.square(members - np.mean(members)))
        sseList.append(SSE)
        
    #get the Index with highest SSE value. THis indicates that the cluster can further be divided into two smaller clusters
    sseArray = np.asarray(sseList)
    dropClusterIndex = np.argsort(sseArray)[-1]
            
    return dropClusterIndex
```

```python
from sklearn.cluster import KMeans
def bisecting_kmeans(matrix, k, iterations):
    
    staringCluster = list();
    clustersList = list();    
    
    for i in range(matrix.shape[0]):
        staringCluster.append(i)
    
    clustersList.append(staringCluster);
    
    #until the number of cluster generated by Bisecting kmeans does not match with the given K value, keep looping
    while len(clustersList) < k:

        dropClusterIndex = calculateSSE(matrix, clustersList)
        droppedCluster = clustersList[dropClusterIndex]
        
        firstCluster, secondCluster = kmeans(matrix[droppedCluster,:], iterations)
        del clustersList[dropClusterIndex]
        
        cluster1 = list();
        cluster2 = list();
        for index in firstCluster:
            cluster1.append(droppedCluster[index])
            
        for index in secondCluster:
            cluster2.append(droppedCluster[index])
        
        clustersList.append(cluster1);
        clustersList.append(cluster2);
    
    labels = [0] * matrix.shape[0];
    
    #allocate clusterID to each document. 
    for index, cluster in enumerate(clustersList):
        for idx in cluster:
            labels[idx] = index + 1
    return labels
```

```python
kValues = list()
scores = list()

for k in range(3, 30, 2):
    labels = bisecting_kmeans(X, k, 10)
    if (k == 7):
         # write result to output file
        outputFile = open("outputFinal.dat", "w")
        for index in labels:
            outputFile.write(str(index) +'\n')
        outputFile.close()
    #compute silhoutte coefficient, since the closest value of Silhoutte coefficent to 1 is for K=7, I have considered 7 as optimal clusters.
    sscore = silhouette_score(X, labels)
    kValues.append(k)
    scores.append(sscore)
    print ("For K= %d silhouette_coefficient Score is %f" %(k, sscore))
    print(labels[0])
 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#Plot a graph for different values of k and their Silhouette scores.
plt.plot(kValues, scores)
plt.xticks(kValues, kValues)
plt.xlabel('Number of Clusters k')
plt.ylabel('silhouette_coefficient Score')
plt.title('Trend of changing silhouette scores for different number of clusters k')
plt.grid(linestyle='dotted')

plt.savefig('plot.png')
plt.show()
```
